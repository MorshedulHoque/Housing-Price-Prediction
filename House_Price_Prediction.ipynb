```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("/content/drive/MyDrive/Datasets/Housing Price/1train.csv")
data.head()
```

```python
data.describe()
```

```python
avg_value = data.LotFrontage.mean()

data['LotFrontage'].fillna(avg_value, inplace = True)
```

```python
data.describe()
```

```python
data.columns
```

```python
features = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
```

```python
X = data[features]
X.head()
```

```python
Y = data.SalePrice
Y.head()
```

```python
train_X, val_X, train_Y, val_Y = train_test_split(X, Y, random_state = 1)
```

**Random forest**

```python
model = RandomForestRegressor(random_state = 1)
model.fit(train_X, train_Y)
```

```python
prediction = model.predict(val_X)
```

```python
mae = mean_absolute_error(val_Y, prediction)
mae
```

```python
rmse = np.sqrt(mean_squared_error(val_Y, prediction))
rmse
```

Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
tree_model = DecisionTreeRegressor(random_state = 1)
tree_model.fit(train_X, train_Y)
tree_prediction = tree_model.predict(val_X)
tree_mae = mean_absolute_error(val_Y, tree_prediction)
```

```python
tree_mae
```

```python
tree_rmse = np.sqrt(mean_squared_error(val_Y, tree_prediction))
tree_rmse
```
